### 5. Feature engineering
#### 5.1 Outliers

```python
corr_df.describe()
```

```python
fig, axes = plt.subplots(nrows=corr_df.shape[1] // 4 + 1, ncols=4, figsize=(20, 5 * (corr_df.shape[1] // 4 + 1)))

for i, column in enumerate(corr_df.columns):
    sns.boxplot(x=corr_df[column], ax=axes[i // 4, i % 4])

plt.tight_layout()
```

We found outliers in cons.conf.idx that i can be replace by mode

```python
descr = corr_df["cons.conf.idx"].describe()
descr
```

```python
minN_iqr = descr["75%"] - descr["25%"]
upper_l = descr["75%"] + 1.5*minN_iqr
lower_l = descr["25%"] - 1.5*minN_iqr

print(lower_l, upper_l)
outliers = corr_df[corr_df['cons.conf.idx'] >= upper_l]
outliers
```

```python
mode = corr_df['cons.conf.idx'].mode()

replace_func = lambda x: x if lower_l <= x <= upper_l else mode
corr_df['cons.conf.idx'] = corr_df['cons.conf.idx'].apply(replace_func)
```

Save all dataframes as csv

```python
df.to_csv('../data/processed/bank-marketing-campaign-data_precleaned.csv', index=False)
corr_df.to_csv('../data/processed/bank-marketing-campaign-data_cleaned.csv', index=False)
outliers.to_csv('../data/processed/bank-marketing-campaign-data_consconfidx-Outliers.csv', index=False)
```

#### 5.3 Feature scaling
First, we divide the data

```python
corr_df.columns
```

```python
predictors = ['age', 'job', 'marital', 'default', 'housing', 'loan', 'contact',
       'month', 'day_of_week', 'duration', 'campaign', 'previous',
       'emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m',
       'nr.employed']
target = 'y'

X = corr_df[predictors]
Y = corr_df[target]

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 42)

X_train.to_csv('../data/processed/bank-marketing-campaign-data_Xtrain_clean.csv', index=False)
X_test.to_csv('../data/processed/bank-marketing-campaign-data_Xtest_clean.csv', index=False)
y_train.to_csv('../data/processed/bank-marketing-campaign-data_ytrain_clean.csv', index=False)
y_test.to_csv('../data/processed/bank-marketing-campaign-data_ytest_clean.csv', index=False)
```

#### 5.4 Normalize

```python
from sklearn.preprocessing import StandardScaler
from pickle import dump

scaler = StandardScaler()
scaler.fit(X_train)

X_train_norm = scaler.transform(X_train)
X_train_norm = pd.DataFrame(X_train_norm, index = X_train.index, columns = predictors)

X_test_norm = scaler.transform(X_test)
X_test_norm = pd.DataFrame(X_test_norm, index = X_test.index, columns = predictors)

dump(scaler, open("../data/processed/normalizer.sav", "wb"))
X_train_norm.to_csv('../data/processed/bank-marketing-campaign-data_norm_Xtrain_clean.csv', index=False)
X_test_norm.to_csv('../data/processed/bank-marketing-campaign-data_norm_Xtest_clean.csv', index=False)

X_train_norm.head()
```

### 6.Feature selection

```python
from sklearn.feature_selection import SelectKBest, mutual_info_classif # mutual_info is used because there are strongly related variables and target is categorical
selection_model = SelectKBest(mutual_info_classif, k = 10)
selection_model.fit(X_train_norm, y_train)
ix = selection_model.get_support()
X_train_sel = pd.DataFrame(selection_model.transform(X_train_norm), columns = X_train_norm.columns.values[ix])

dump(selection_model, open("../data/processed/selection_model.sav", "wb"))
X_train_sel.to_csv('../data/processed/bank-marketing-campaign-data_norm_Xtrain_clean-featureSel.csv', index=False)

X_train_sel.head()
```
